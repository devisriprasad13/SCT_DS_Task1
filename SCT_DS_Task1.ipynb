```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import os

# Step 2: Load the population data CSV
df = pd.read_csv("/content/drive/MyDrive/Task1/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Step 3: Filter for India and extract year-wise population
india_df = df[df["Country Name"] == "India"]
years = [str(year) for year in range(1990, 2023)]
population = india_df[years].T  # Transpose for plotting
population.columns = ["Population"]
population.index.name = "Year"
population.reset_index(inplace=True)

# Step 4: Plot as a bar chart
plt.figure(figsize=(14, 6))
plt.bar(population["Year"], population["Population"], color="skyblue")
plt.title("India's Total Population (1990–2022)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
